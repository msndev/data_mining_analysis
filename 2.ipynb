```python
print ("Elias gamma encoding of even numbers from 1 to 20 18BCE0559, Sreenitya Mandava")

from math import log

log2 = lambda x: log(x, 2)
s = {}

def Unary(x):
  return (x-1)*'0'+'1'

def Binary(x, l = 1):
  s = '{0:0%db}' % l
  return s.format(x)

def Elias_Gamma(x):
  if(x == 0):
    return '0'

  n = 1 + int(log2(x))
  b = x - 2**(int(log2(x)))

  l = int(log2(x))

  print('Selector = ',Unary(n), end =' and ')
  print('Offset = ',Binary(b, l))
  print('Code = ',Unary(n) + Binary(b, l))
x = 2
for x in range(1, 21):
  if x % 2 == 0:
    print('\n',x,'--->')
    Elias_Gamma(x)
```

```python
print ("Elias delta encoding of even numbers from 1 to 20 18BCE0559, Sreenitya Mandava")

from math import log

log2 = lambda x: log(x, 2)

def Selector(x, y):
  x = (x-1)*'0'+'1'
  t = '{0:0%db}' % 1
  k = t.format(y)
  k.replace(k[0],'')
  return x + k

def Offset(x, l = 1):
  s = '{0:0%db}' % l
  t = s.format(x)
  return t.replace(t[0],'')

def Elias_Delta(x):
  if(x == 0):
    return '0'

  m = 1 + int(log2(x))
  n = 1 + int(log2(m))

  l = int(log2(x))

  print('Selector = ',Selector(n, m), end = ' and ')
  print('Offset = ',Offset(x, l))
  print('Code = ',Selector(n, m) + Offset(x, 1))

x = 2
for x in range(1, 21):
  if x % 2 == 0:
    print('\n',x,'--->')
    Elias_Delta(x)
```

```python
print("Golomb encoding of even numbers from 1 to 20 and b = 10 18BCE0559, Sreenitya Mandava")

from math import log

log2 = lambda x: log(x, 2)

x = 2
b = 10

for x in range(1, 21):
  if x % 2 == 0:
    print('\n',x,'--->')
    q = x//b
    r = x - q * b
    quo ='0'*q+'1'

    i = int(log2(b))
    d = 2**(i+ 1)-b

    if r < d:
      rem = bin(r)[2:]
      l = len(rem)

      if l<b:
        rem = '0'*(i-l)+rem
    else:
      rem = bin(r + d)[2:]
      l = len(rem)

      if l < i + 1:
        rem = '0'*(i + 1-l)+rem
    print ('Quotient = ',quo, end = ' and ')
    print ('Remainder = ',rem)
    print ('Code = ',quo + rem)
```
